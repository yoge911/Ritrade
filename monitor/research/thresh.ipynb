```python
import pandas as pd

# Load your collected baseline CSV
data = pd.read_csv('btcusdt_baseline_20250427_020840.csv')  # Change filename if needed

# ----------------
# Volume Analysis
# ----------------
total_volume = data['total_volume'].sum()
volume_median = data['total_volume'].median()
volume_percentile_80 = data['total_volume'].quantile(0.8)
volume_percentile_20 = data['total_volume'].quantile(0.2)

# ----------------
# Trade Count Analysis
# ----------------
trade_median = data['trade_count'].median()
trade_percentile_80 = data['trade_count'].quantile(0.8)
trade_percentile_20 = data['trade_count'].quantile(0.2)

# ----------------
# Std Dev Analysis
# ----------------
std_median = data['std_dev'].median()
std_percentile_80 = data['std_dev'].quantile(0.8)
std_percentile_20 = data['std_dev'].quantile(0.2)

# ----------------
# Print everything nicely
# ----------------
print("\n📊 Analysis of Baseline Data")
print("-----------------------------------")
print(f"Total Volume: {round(total_volume, 4)}")
print(f"Median Volume: {round(volume_median, 4)}")
print(f"Volume 80th Percentile (high activity): {round(volume_percentile_80, 4)}")
print(f"Volume 20th Percentile (low activity): {round(volume_percentile_20, 4)}")
print("-----------------------------------")
print(f"Median Trade Count: {round(trade_median, 2)}")
print(f"Trade Count 80th Percentile: {round(trade_percentile_80, 2)}")
print(f"Trade Count 20th Percentile: {round(trade_percentile_20, 2)}")
print("-----------------------------------")
print(f"Median Std Dev: {round(std_median, 6)}")
print(f"Std Dev 80th Percentile: {round(std_percentile_80, 6)}")
print(f"Std Dev 20th Percentile: {round(std_percentile_20, 6)}")
print("-----------------------------------")
```
